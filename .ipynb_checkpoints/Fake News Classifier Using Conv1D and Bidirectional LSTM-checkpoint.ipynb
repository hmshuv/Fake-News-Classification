## Fake News Classifier Using Bidirectional LSTM

Dataset: https://www.kaggle.com/c/fake-news/data#

```python
import pandas as pd
```

```python
df=pd.read_csv('data/train.csv')
```

### Data Preprocessing

```python
df.head()
```

```python
df=df.dropna()
```

```python
X=df.drop('label',axis=1)
```

```python
y=df['label']
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import *
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from tensorflow.keras.callbacks import EarlyStopping
```

```python
voc_size=5000
```

### Onehot Representation

```python
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

### Embedding Representation

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
def define_model(dp=0.1,voc_size=5000,sent_length=20,embedding_vector_features=40,
                 l1=100):
    embedding_vector_features=40
    model1=Sequential()
    model1.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
    model1.add(Dropout(dp))
    model1.add(Conv1D(64,3,activation='relu',padding='same'))
    model1.add(MaxPooling1D(2,2))
    model1.add(Dropout(dp))
    model1.add(Conv1D(32,3,activation='relu',padding='same'))
    model1.add(MaxPooling1D(2,2))
    model1.add(Dropout(dp))
    model1.add(Bidirectional(LSTM(l1)))
    model1.add(Dropout(dp))
    model1.add(Dense(1,activation='sigmoid'))
    model1.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
    print(model1.summary())
    return model1
```

```python
dp = [0.1,0.3,0.6]
l1 = [75,100]
batch_size=[1,16,32]
epochs=[5]
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
model = KerasClassifier(build_fn=define_model,verbose=1)
```

```python
param_grid = dict(dp = dp,l1=l1,batch_size=batch_size,epochs=epochs) 
```

# Train-Test Split

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

### Grid Search

```python
bilstm_gridcv = GridSearchCV(estimator = model,param_grid=param_grid,cv=3,verbose=2,n_jobs=-1)
bilstm_gridcv.fit(X_train,y_train)
```

```python
bilstm_gridcv.best_params_
```

```python
embedding_vector_features=40
model1=Sequential()
model1.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model1.add(Dropout(0.6))
model1.add(Conv1D(64,3,activation='relu',padding='same'))
model1.add(MaxPooling1D(2,2))
model1.add(Dropout(0.6))
model1.add(Conv1D(32,3,activation='relu',padding='same'))
model1.add(MaxPooling1D(2,2))
model1.add(Dropout(0.6))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dropout(0.6))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
print(model1.summary())
```

### Model Training

```python
model1.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=5,batch_size=32,verbose=1)
```

```python
model_loss = pd.DataFrame(model1.history.history)
model_loss.plot()
```

### Performance Metrics And Accuracy

```python
y_pred1 = (model1.predict(X_test) > 0.5).astype("int32")
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
